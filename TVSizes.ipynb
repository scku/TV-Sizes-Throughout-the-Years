# TV Sizes Throughout the Years
Using data collected from CNET, I analyzed the data to find the trend in TV sizes over the years.

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
print "Shape: {0}".format(df.shape)
print "Columns: {0}".format(df.columns.values)
```

```python
print df.head()
print df.tail()
```

## Preprocessing
Clean the data and remove any entries with NaN/Null values

```python
# Convert review_date to datetime objects
MONTHS = {'Jan.':'January', 'Feb.':'February', 'Mar.':'March', 'Apr.':'April',
          'May.':'May', 'Jun.':'June', 'Jul.':'July', 'Aug.':'August', 'Sep.':'September',
          'Oct.':'October', 'Nov.':'November', 'Dec.':'December'}
def convertDate(x):
    try:
        m, d, y = x.split()
        m = MONTHS[m]
        return ' '.join([m, d, y])
    except:
        return x

# Function to check if date is NaN
def isDateNaN(x):
    try:
        return np.isnan(x)
    except:
        return False
    
# Change diagonal_size column to float values
def toNumber(x):
    try:
        return float(re.sub(r'[^0-9.]', '', x))
    except:
        return x    
    
df['diagonal_size'] = df['diagonal_size'].apply(toNumber)
df['review_date'] = df['review_date'].apply(convertDate)

# Find number of rows with NaN/Null diagonal_size entry
index_size = df['diagonal_size'].index[df['diagonal_size'].apply(np.isnan)]
index_date = df['review_date'].index[df['review_date'].apply(isDateNaN)]
print "Number of NaN diagonal_size entries: {0}".format(len(index_size))
print "Number of NaN review_date entries: {0}".format(len(index_date))

# Remove rows with NaN value
index = index_size + index_date
df_subset = df.drop(index)
print df_subset.shape

# Convert review_date to datetime object
df_subset['review_date_obj'] = pd.to_datetime(df_subset['review_date'], format="%B %d, %Y")
```

##Analyze

```python
# Groupby year
gb = df_subset.groupby(df_subset['review_date_obj'].map(lambda x: x.year))
#print gb['diagonal_size'].apply(lambda x: x.describe())

YEARS = [2008, 2009, 2010, 2011, 2012, 2013, 2014]
mean = []
median = []

plt.figure(figsize=(8,35))
for (i, obj) in enumerate(gb):
    year, group = obj
    if year in YEARS:
        cur_mean = group['diagonal_size'].mean()
        cur_median = group['diagonal_size'].median()
        mean.append(cur_mean)
        median.append(cur_median)
        print "{0} diagonal size - median: {1}  mean: {2}".format(year, cur_median, cur_mean)
        plt.subplot(7,1,i-5)        
        plt.hist(list(group['diagonal_size'].values), align='mid')
        plt.xlim(xmin=5, xmax=85)
        plt.ylim(ymax=25, ymin=0)
        # Labels
        plt.xlabel('Diagonal Size')
        plt.ylabel('Count')
        plt.title('TV sizes in {0}'.format(year))
plt.show()
```

```python
# Plot median vs year
plt.figure(figsize=(10,6))
plt.plot(YEARS, median, 'b')
plt.plot(YEARS, mean, 'r')
plt.xlim(2008,2014)
plt.ticklabel_format(style='plain', axis='x', useOffset=False)
plt.legend(('Median', 'Mean'), loc='upper left')
plt.xlabel('Year')
plt.ylabel('Size')
plt.title('Median Diagonal Size vs Year')
plt.show()
```
